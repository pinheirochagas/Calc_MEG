```python
######################################################## Needed tools for ANALYSES
import numpy as np
import sys as sys
import nibabel as nb
import scipy.io as io
from scipy import stats

from sklearn import linear_model,metrics,svm
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import permutation_test_score, LeaveOneOut, KFold, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectPercentile, SelectKBest, f_classif
from sklearn.metrics import precision_score

import warnings
warnings.filterwarnings('ignore')

from sklearn.externals.joblib import Memory,Parallel, delayed
mem = Memory('/volatile/tmp')
_svc.score = mem.cache(svc.score)
_kendall_score = mem.cache(svc.score)

######################################################## Needed tools for PLOTTING
%pylab inline 
import prettyplotlib as ppl
import matplotlib.pyplot as plt
from prettyplotlib import brewer2mpl
import matplotlib.gridspec as gridspec

# More colors and right font
ppl.colors.set4 = ppl.colors.set2 + ppl.colors.set2 + ppl.colors.set2
font = {'fontname':'Century Gothic'}
```

```python
# ================================== Subjects
SUBJECTS = ['s03','s05','s06','s08','s09','s10','s11','s12','s14','s15']
```

```python
# Metric for regression = kendall_tau
def kendall_score(_x, _y):
    _x = np.array(_x).astype(np.float)
    _y = np.array(_y).astype(np.float)
    out = stats.kendalltau(_x, _y)[0]
    if np.isnan(out):
        out = 0.
    return out
```

```python
# ================================== TRAIN ON SMALL TEST ON BIG ==================================

# Loop across subjects
AllScores_class= np.zeros(shape=(10,1001))
AllScores_regr= np.zeros(shape=(10,1001))
sub = 0
for s in SUBJECTS:
    print 'Loading %s' % s
    
    # ================================== Load data
    data = []
    data = io.loadmat('data/%s.mat' % s, matlab_compatible=True)
    D = data['D'][0]
    # keep needed ones
    Dev8 = D['Dev8'][0]
    Dev10 = D['Dev10'][0]
    Dev13 = D['Dev13'][0]
    Dev20 = D['Dev20'][0]
    Dev25 = D['Dev25'][0]
    Dev32 = D['Dev32'][0]
    # delete in order to save memory
    data = []
    D = []
    
    # ================================== Left side of the slope - ie small number
    # Data
    Left = []
    Left = np.concatenate((Dev8,Dev10, Dev13), axis=0)
    # Labels
    D8=[0]*Dev8.shape[0]
    D10=[1]*Dev10.shape[0]
    D13=[2]*Dev13.shape[0]
    D8 = np.array(D8)
    D10 = np.array(D10)
    D13 = np.array(D13)
    LLabel = np.concatenate((D8,D10, D13), axis=0)
    
    # ================================== Right side of the slope - ie big number
    # Data
    Right = []
    Right = np.concatenate((Dev20,Dev25, Dev32), axis=0)
    # Labels
    D20=[2]*Dev20.shape[0]
    D25=[1]*Dev25.shape[0]
    D32=[0]*Dev32.shape[0]
    D20 = np.array(D20)
    D25 = np.array(D25)
    D32 = np.array(D32)
    RLabel = np.concatenate((D20,D25, D32), axis=0)    
    
    # ================================== Decoding tools
    # Create scaler
    scaler= StandardScaler()
    # Create classifier object 
    svc = svm.SVC()
    # Create linear regression 
    reg = linear_model.RidgeCV()
    
    # ================================== Loop in diagonal
    y = LLabel
    y2 = RLabel
    
    print 'Classification for %s' % s
    scores_class = []
    for tp in range(0,1001):
        P = []
        X = Left[:,:,tp]
        X2 = Right[:,:,tp]
        # normalization
        scaler.fit(X)
        X = scaler.transform(X)
        #scaler.fit(X2)
        X2 = scaler.transform(X2)
        # training
        svc.fit(X,y)
        # test
        scores_class.append(svc.score(X2,y2))
        
    # plot all scores    
    scores_class = np.array(scores_class)
    plt.plot(scores_class,color=ppl.colors.set4[0])
    plt.show()    
    
    print 'Regression for %s' % s
    scores_reg = []
    for tp in range(0,1001):
        P = []
        X = Left[:,:,tp]
        X2 = Right[:,:,tp]
        # normalization
        scaler.fit(X)
        X = scaler.transform(X)
        #scaler.fit(X2)
        X2 = scaler.transform(X2)
        reg.fit(X,y)
        P = reg.predict(X2)
        #scores.append(reg.score(X2,y2))
        scores_reg.append(kendall_score(P,y2))
        
    # plot all scores    
    scores_reg = np.array(scores_reg)
    plt.plot(scores_reg,color=ppl.colors.set4[3])
    plt.show()    
    
    print ''
    print ''
    
    AllScores_class[sub]= scores_class
    AllScores_regr[sub]= scores_reg
    
    sub = sub + 1

# Compute and Plot mean and error
print 'Mean Classification'
Mean_class = np.mean(AllScores_class,0)
t = np.arange(len(Mean_class))
Std_class = np.std(AllScores_class,0)
Error_class = Std_class/np.sqrt(3)
plt.plot(t,Mean_class,color=ppl.colors.set4[0])
plt.fill_between(t,Mean_class-Error_class, Mean_class+Error_class,
    alpha=0.5, edgecolor=ppl.colors.set4[4], facecolor=ppl.colors.set4[4])
plt.show()

print 'Mean Regression'
Mean_regr = np.mean(AllScores_regr,0)
Std_regr = np.std(AllScores_regr,0)
Error_regr = Std_regr/np.sqrt(10)
plt.plot(t,Mean_regr,color=ppl.colors.set4[6])
plt.fill_between(t,Mean_regr-Error_regr, Mean_regr+Error_regr,
    alpha=0.5, edgecolor=ppl.colors.set4[5], facecolor=ppl.colors.set4[5])
plt.show()     
```
